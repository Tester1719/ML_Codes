```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from hmmlearn.hmm import GaussianHMM
from sklearn.datasets import load_iris
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler , StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.metrics import confusion_matrix , classification_report , precision_score , recall_score, f1_score, roc_curve , auc , accuracy_score
```

```python
iris_data = load_iris()
```

```python
x = iris_data.data
y = iris_data.target
```

```python
def plot_confusion_matrix(cm , title):
  plt.figure(figsize=(3,4))
  sns.heatmap(cm)
  plt.xlabel("Predicted_labels")
  plt.ylabel("True_labels")
  plt.title(title)
  plt.show()
```

```python
def evaluate_model(y_test,y_pred):
  cm = confusion_matrix(y_test,y_pred)
  print('Confusion Matrix is \n')
  plot_confusion_matrix(cm , 'Confusion Matrix')
  print('\n')
  clr = classification_report(y_test,y_pred)
  print('classification report is \n')
  print(clr)
  print('\n')
  accuracy = accuracy_score(y_test,y_pred)
  print(f"Accuracy Score is - {accuracy*100}")
```

```python
from hmmlearn.hmm import GaussianHMM, MultinomialHMM
from sklearn.metrics import accuracy_score , classification_report , confusion_matrix
```

```python
def prepare_data(x,y,scaler,test_size):
  x_train, x_test, y_train, y_test  = train_test_split(x, y, test_size=test_size, random_state=42)
  x_train_scaled = scaler.fit_transform(x_train)
  x_test_scaled = scaler.fit_transform(x_test)
  return x_train_scaled , x_test_scaled , y_train , y_test
```

```python
test_sizes = [0.35 , 0.3 , 0.2]
results = []

for test_size in test_sizes:
  print(f"Evaluating with the test size - {test_size}\n")
  print("GaussianHMM Results ####### \n")
  x_train, x_test, y_train, y_test  = train_test_split(x,y,test_size = test_size , random_state = 42)
  model = GaussianHMM(n_components=2, covariance_type="diag", n_iter=1000)
  model.fit(x_train)
  y_pred_gaussian = model.predict(x_test)
  evaluate_model(y_test,y_pred_gaussian)
  accuracy_gaussian = accuracy_score(y_test,y_pred_gaussian)
  precision_gaussian = precision_score(y_test,y_pred_gaussian,average='weighted')
  recall_gaussian = recall_score(y_test,y_pred_gaussian,average='weighted')
  f1_gaussian = f1_score(y_test,y_pred_gaussian,average='weighted')
  results.append({"model" : "GaussianHMM" ,
                  "test_size" : test_size ,
                  "accuracy" : accuracy_gaussian,
                  "precision" : precision_gaussian,
                  "recall" : recall_gaussian,
                  "f1_score" : f1_gaussian})
  print('\n')
  print("MultinomialHMM Results - #########\n")
  scaler = MinMaxScaler()
  x_train_multi , x_test_multi , y_train_multi , y_test_multi = prepare_data(x,y,scaler,test_size)
  model = MultinomialHMM(n_components=2,n_iter=1000)
  x_train_multi = np.round(x_train_multi).astype(int)
  x_test_multi = np.round(x_test_multi).astype(int)
  model.fit(x_train_multi)
  y_pred_multinomial = model.predict(x_test_multi)
  evaluate_model(y_test_multi,y_pred_multinomial)
  accuracy_multinomial = accuracy_score(y_test,y_pred_multinomial)
  precision_multinomial = precision_score(y_test,y_pred_multinomial,average='weighted')
  recall_multinomial = recall_score(y_test,y_pred_multinomial,average='weighted')
  f1_multinomial = f1_score(y_test,y_pred_multinomial,average='weighted')
  results.append({"model" : "MultinomialHMM",
                  "test_size" : test_size,
                  "accuracy" : accuracy_multinomial,
                  "precision" : precision_multinomial,
                  "recall" : recall_multinomial,
                  "f1_score" : f1_multinomial})
  print('\n')
  print('\n\n')



```
